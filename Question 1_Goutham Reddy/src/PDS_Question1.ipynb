```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("/content/drive/MyDrive/Question 1/Data_Clean/Question1CSV.csv")

```

```python
plt.hist(data["Grip strength"], bins=10)
plt.xlabel("Grip Strength")
plt.ylabel("Count")
plt.title("Distribution of Grip Strength")
plt.savefig("/content/drive/MyDrive/Question 1/results/img1.jpg")
```

```python
sns.boxplot(data=data, x="Age", y="Grip strength")
plt.title("Age vs Grip Strength")
plt.savefig("/content/drive/MyDrive/Question 1/results/img2.jpg")
```

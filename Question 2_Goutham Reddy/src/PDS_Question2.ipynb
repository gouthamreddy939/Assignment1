```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Question 2/data_clean/StudentsPerformance.csv")
```

```python
print(data.head())
```

```python
plt.hist(data["math score"], bins=20)
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.savefig("/content/Question 2/results/img1.jpg")
```

```python
sns.scatterplot(data=data, x="math score", y="reading score")
plt.savefig("/content/Question 2/results/img2.jpg")
```

```python
sns.countplot(data=data, x="parental level of education")
plt.xticks(rotation=45)
plt.savefig("/content/Question 2/results/img3.jpg")
```

```python
sns.boxplot(data=data, x="gender", y="math score", hue="test preparation course")
plt.savefig("/content/Question 2/results/img4.jpg")
```

```python
ct = pd.crosstab(data["lunch"], data["parental level of education"])
ct.plot(kind="bar", stacked=True)
plt.xlabel("Lunch")
plt.ylabel("Count")
plt.savefig("/content/Question 2/results/img5.jpg")
```
